```python
# Importing the necessary libraries

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
fetch_california_housing()
```

```python
data = fetch_california_housing(as_frame=True).frame
data
```

```python
data.dtypes
```

```python
# OR

# df = fetch_california_housing()
# df
```

```python
# dataset = pd.DataFrame(df.data)
# dataset.columns = df.feature_names
```

```python
data.head()
```

```python
target_data = data['MedHouseVal']
target_data
```

```python
data_ = data.drop(columns=["MedHouseVal"])
data_
```

```python
data_.columns
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data_, target_data, random_state=42, test_size=0.25)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.model_selection import cross_validate
cv_result_mse = cross_validate(model, X_train, y_train, scoring='neg_mean_squared_error', cv=5)
cv_result_mse
```

```python
# from sklearn.model_selection import cross_validate
# cv_result = cross_validate(model, X_train, y_train, cv=5)
```

```python
# cv_result
```

```python
scores = cv_result_mse['test_score']
scores
```

```python
print(f"The mean cross validation accuracy is : {scores.mean():.3f} ± {scores.std():.3f}")
```

```python
# scores.mean()
# np.mean(scores)
```

```python
# Prediction

reg_pred = model.predict(X_test)
reg_pred[:5]
```

```python
reg_pred
```

```python
y_test.to_numpy()
```

```python
import seaborn as sns
# sns.displot(reg_pred - y_test, kind='kde')
sns.displot(reg_pred - y_test.to_numpy(), kind='kde')
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test, reg_pred)

score
```
